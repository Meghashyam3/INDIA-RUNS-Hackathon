# Step 1.4: Data Flattening & Exploration

In this notebook, we verify our data ingestion script and use Pandas to flatten the deeply nested candidate JSON (like the `profile`, `career_history`, and `redrob_signals` objects) into a manageable tabular format.

```python
import sys
import pandas as pd
import json

# Add the src directory to Python's path so we can import our script
sys.path.append('../src')
from data_ingestion import load_candidates

# Load the sample candidates
sample_path = '../[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/sample_candidates.json'
candidates_raw = load_candidates(sample_path)
print(f"Loaded {len(candidates_raw)} records.")
```

### Flattening the JSON

The raw JSON has nested dictionaries. We can use `pd.json_normalize` to flatten them cleanly. For example, `profile.current_title` will become its own column.

```python
# Flatten the core properties and nested dicts (profile, redrob_signals)
df = pd.json_normalize(candidates_raw)

# Let's inspect all the columns created from flattening
print("Total Columns:", len(df.columns))
print("\nSample columns:")
for col in list(df.columns)[:15]:
    print(" -", col)

# View the first 3 rows
df.head(3)
```

### Handling Arrays (Career History & Skills)

Notice that `career_history` and `skills` are still lists of dictionaries inside the DataFrame. In **Phase 2**, we will write a `Candidate Stringifier` to convert these arrays into a single clean text string for embedding!

```python
# Previewing how nested lists look inside a single cell
candidate_zero_skills = df.loc[0, 'skills']
print("Skills array for Candidate 1:\n")
print(json.dumps(candidate_zero_skills, indent=2))
```
