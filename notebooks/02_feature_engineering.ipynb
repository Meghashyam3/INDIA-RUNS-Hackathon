# Step 2.1 & 2.2: The Candidate Stringifier

To use semantic search (Vector Embeddings), we must convert each candidate's deeply nested JSON profile into a single block of text that represents their "professional essence". 

We have built `stringify_candidate()` in `src/features.py`. Let's test it on our sample dataset.

```python
import sys
import json
sys.path.append('../src')

from data_ingestion import load_candidates
from features import stringify_candidate

# Load data
sample_path = '../[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/sample_candidates.json'
candidates = load_candidates(sample_path)
print(f"Loaded {len(candidates)} candidates.\n")
```

### Let's see the stringified output for the first candidate:

```python
candidate_zero = candidates[0]
stringified_text = stringify_candidate(candidate_zero)

print("RAW STRINGIFIED OUTPUT:\n")
print(stringified_text)
print("\n---\nCharacter length:", len(stringified_text))
```

### Apply to all candidates
Let's stringify all 50 sample candidates and see if it breaks.

```python
all_strings = [stringify_candidate(c) for c in candidates]
print(f"Successfully stringified {len(all_strings)} candidates.")
print("Average length:", sum(len(s) for s in all_strings) / len(all_strings), "characters.")
```
